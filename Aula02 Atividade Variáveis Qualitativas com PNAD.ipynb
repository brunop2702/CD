```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_excel("PNAD 2013 A.xlsx")
```

```python
dados.head()
```

```python
n=dados.shape[0]
```

```python
dados.V0302=dados.V0302.astype('category')
dados.V0404=dados.V0404.astype('category')
dados.V6002=dados.V6002.astype('category')
dados.V4802=dados.V4802.astype('category')
dados.V0601=dados.V0601.astype('category')
dados.V6003=dados.V6003.astype('category')
```

```python
#tabelas univariadas
print('tabela frequencia relativa (Sexo)')
dados.V0302.cat.categories = (['Masculino','Feminino'])
ut1=(dados.V0302.value_counts()/n*100).round(decimals=1)
print(ut1,'\n')

print('tabela frequencia relativa (cor ou raça)')
dados.V0404.cat.categories = (['Amarela','Branca','Preta','Indígena','Parda','Sem declaração'])
ut2=(dados.V0404.value_counts()/n*100).round(decimals=1)
print(ut2,'\n')

print('tabela frequencia relativa (rede de ensino)')
dados.V6002.cat.categories = (['Pública','Privada'])
ut3=(dados.V6002.value_counts()/n*100).round(decimals=1).reindex(['Pública','Privada','NA'],fill_value=100-21.5-6.9)
print(ut3,'\n')

print('tabela frequencia relativa nível de ensino e grupos de séries do ensino fundamental que frequentavam (todos os estudantes)')
dados.V4802.cat.categories = (['maternal/jardim de infancia','classe de alfabetizção','alfabetização de adultos'
                              ,'1 a 4 series EF em 8 anos','5 a 8 series EF em 8 anos','1 serie EF em 9 anos'
                              ,'2 a 5 series EF em 9 anos','6 a 9 series EF em 9 anos','educação de jovens e adultos'
                              ,'ensino medio','ensino superior'])
ut4=(dados.V4802.value_counts()/n*100).round(decimals=1)
print(ut4,'\n')

print('tabela frequência relativa (sabe ou não ler e escrever)')
dados.V0601.cat.categories = (['sim','não'])
ut5=(dados.V0601.value_counts()/n*100).round(decimals=1)
print(ut5,'\n')

print('tabela frequência relativa (curso que frequenta)')
dados.V6003.cat.categories = (['regular do ensino fundamental','regular do ensino medio','Educação de jovens e adultos ou supletivo do ensino fundamental'
                              ,'Educação de jovens e adultos ou supletivo do ensino médio','superior de graduação'
                              ,'alfabetização de jovens e adultos','creche','classe de alfabetizção','maternal / jardim de infancia'
                              ,'pré-vestibular','mestrado ou doutorado'])
ut6=(dados.V6003.value_counts()/n*100).round(decimals=1)
print(ut6,'\n')

print('tabela frequencia relativa (regiões)')
faixa=range(10,70,10)
dados.UF2=pd.cut(dados.UF,faixa)
dados.UF2.cat.categories=(['norte','nordeste','sudeste','sul','centroeste'])
dados.UF2.value_counts(True)*100
ut8=dados.UF2.value_counts(True)*100
print(ut8)
```
